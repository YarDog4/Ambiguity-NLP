```python
from transformers import pipeline

# In this code below, I'm loading in a pre-trained model and it's attached tokenizer from HuggingFace.
# Load the ambiguity detection model

classifier = pipeline(

    "text-classification",

    #model="j-hartmann/ambiguity-distilroberta-base",

    model = "Slomb/Ambig_Question",

    #model = "redphoenix1996/srs_ambiguity_generation_1", # this model is not free so I'm unable to access it.

    #tokenizer="j-hartmann/ambiguity-distilroberta-base"

    tokenizer= "Slomb/Ambig_Question",

    #tokenizer = "redphoenix1996/srs_ambiguity_generation_1", # this model is not free so I'm unable to access it.
)

# Test texts

texts = [

    "Did he do it?",

    "The book is on the table.",

    "They saw her duck.",

    "What time does the train arrive?"

] # We can chnage this sentence later if it's too simple.

# Run predictions

for t in texts:

    result = classifier(t)[0]

    print(f"Text: {t}")

    print(f" → Label: {result['label']}, score: {result['score']:.4f}")

    print()

```
